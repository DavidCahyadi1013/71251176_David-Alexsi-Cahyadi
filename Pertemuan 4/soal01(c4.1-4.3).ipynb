```python
#Contoh 4.1
isuhu = input("Masukkan suhu tubuh: ")

try: 
   suhu = int(isuhu)
   
   if suhu >= 38:
    	print("Anda demam")
   else:
    	print("Anda tidak demam")

except:
   print("Input yang dimasukkan tidak valid")
```

```python
#Contoh 4.2
angka = input("Masukan Suatu Bilangan: ")

try: 
    bilangan = int(angka)

    if bilangan > 0:
	    print ("Positif")
    elif bilangan < 0:
	    print ("Negatif")
    elif bilangan == 0:
	    print ("Nol")
	
except:
	print("Input yang dimasukkan tidak valid")
```

```python
#Contoh 4.3
a1 = input("Masukkan bilangan pertama: ")
b1 = input("Masukkan bilangan kedua: ")
c1 = input("Masukkan bilangan ketiga: ")
try:
    a = int(a1)
    b = int(b1)
    c = int(c1)

    if a > b and a > c:
    	print("Terbesar:", a)
    elif b > a and b > c:
        print("Terbesar:", b)
    elif c > a and c > b:
        print("Terbesar:", c)

except:
    print("Input yang dimasukkan tidak valid")
```
